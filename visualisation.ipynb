```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris.head(3)
```

```python
iris['species'].unique()
```

```python
dummy={'setosa':0,'versicolor':1,'virginica':2}
```

```python
iris['species']=iris['species'].map(dummy)
iris.head(2)
```

```python
X=iris.drop('species',axis=1)
X
```

```python
comp=decomposition.PCA(n_components=3)
comp.fit(X)
X=comp.transform(X)
y=iris['species']
```

```python
fig=plt.figure(figsize=(10,10))
from mpl_toolkits.mplot3d import Axes3D
a=Axes3D(fig)
a.scatter(X[:,0],X[:,1],X[:,2],c=y,cmap='viridis')
plt.show()
```
